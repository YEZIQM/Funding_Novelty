```python
import pandas as pd
```

```python
def trans_out_table(df_div):
    df = df_div.copy()
    
    df['CI_95_lower'] = df['CI_lower'].div(df['mean_ctl'])*100
    df['CI_95_higher'] = df['CI_higher'].div(df['mean_ctl'])*100
    df['delta'] = df['beta'].div(df['mean_ctl'])*100
    df['delta']  = df['delta'].round(2)
    df.index = df['Unnamed: 0']
    df['CI_95_lower'] = df['CI_95_lower'].round(2).astype(str)
    df['CI_95_higher'] = df['CI_95_higher'].round(2).astype(str)
    m = list(df[['CI_95_lower','CI_95_higher']].values)
    li = [list(a) for a in m]
    df['CI_95'] = li
    df.index.name = None
    
    return df[['num_C',
       'num_T', 'num_C_matched', 'num_T_matched', 'delta', 'CI_95','p']]
```

```python
df_PA = pd.read_csv('/data/home/bsw639/Expertise/notebooks/R_out/res_PA.csv')
```

```python
df_div = trans_out_table(df_PA)
```
